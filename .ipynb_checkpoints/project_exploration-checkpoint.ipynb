```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import cv2

def gauss(mu, sigma):
    def f(x):
        return torch.exp(-(mu-x)**2 / (2*sigma**2))
    return f

def get_Y(
    alpha, stride, phis,
    a, mu, sigma,
    plot=False, colors=None,
):
    if len(phis.shape) == 2:
        phis = phis.repeat([a.shape[0], 1, 1])

    if colors is None:
        colors = ["b" for _ in range(phis.shape[1])]

    n, n_philters, philter_len = phis.shape
    N = int(np.ceil(1 + alpha * philter_len * (1 + (n-1)*stride)))
    
    if plot:
        fig = plt.figure()
        ax = fig.gca()
        ax.set_xlabel("x")
        ax.set_ylabel("y")

        X = torch.arange(N)

    Y = torch.zeros(N)

    for i in range(n):
        for k in range(philter_len):
            for channel in range(n_philters):
                offset = -sigma[i, channel]*alpha*(philter_len/2 - (k+0.5))
                this_mu = alpha*philter_len*stride*(i + mu[i, channel]) \
                    + alpha*philter_len/2 \
                    + offset
                this_sigma = sigma[i, channel]*alpha/2
                this_gauss = gauss(this_mu, this_sigma)

                min_x = int(max(0, this_mu.detach() -3*this_sigma.detach()))
                max_x = int(min(N, this_mu.detach() +3*this_sigma.detach()))

                this_Y = torch.zeros_like(Y)
                for j in range(min_x, max_x):
                    this_Y[j] = a[i, channel] * phis[i, channel, k] * this_gauss(j)
                Y += this_Y

                if plot:
                    ax.plot(X, this_Y.detach(), color=colors[channel])

    if plot:
        ax.plot(X, Y.detach(), color="black")
        plt.show()

    return Y

def viz_phis(phis, alpha, colors=None):
    if len(phis.shape) == 2:
        phis = torch.unsqueeze(phis, 0)
        
    n, n_philters, philter_len = phis.shape

    if colors is None:
        colors = ["b" for _ in range(n_philters)]

    fig = plt.figure(figsize=(2*n, 2*2*n_philters))
    
    for i in range(n):
        for j in range(n_philters):
            ax = fig.add_subplot(2*n_philters, n, 2*n*j+i+1)
            ax.imshow(np.array(phis[i, j]).reshape(1, -1), cmap="gray", vmin=0, vmax=1)
            ax.set_title("(x, ch) = ({}, {})".format(i, j))

            ax = fig.add_subplot(2*n_philters, n, 2*n*j+n+i+1)
            Y = get_Y(alpha, stride=1, phis=phis[i, j].reshape(1, 1, -1),
                     a=torch.ones(1, 1), mu=torch.zeros(1, 1), sigma=torch.ones(1, 1))
            ax.plot(Y.detach(), color=colors[j])

    plt.show()

def plot_coefs(a, mu, log_sigma):
    if len(a.shape) == 2:
        a = torch.unsqueeze(a, 0)
        mu = torch.unsqueeze(mu, 0)
        log_sigma = torch.unsqueeze(log_sigma, 0)

    coef_names = ["a", "mu", "log_sigma"]
    coef_colors = ['g', 'r', 'b']

    t, n, n_philters = a.shape

    if t > 1:
        try:
            os.system("rm plot_coefs_*.png")
        except:
            pass
        try:
            os.system("rm plot_coefs.mp4")
        except:
            pass
    
    for k in range(t):
        fig = plt.figure(figsize=(2*n, 2*n_philters))
        plt.xlim(0, n)
        plt.ylim(0, n_philters)
        plt.xlabel("x")
        plt.ylabel("ch")

        for i in range(1, n):
            plt.plot([i, i], [0, n_philters], color="black", linewidth=5.0)

        max_a = a[k].abs().max()
        max_mu = mu[k].abs().max()
        max_log_sigma = log_sigma[k].abs().max()
        
        for i in range(n):
            for j in range(n_philters):
                ax = fig.add_subplot(n_philters, n, j*n+i+1)
                ax.set_xlim(-1, 3)
                ax.set_ylim(-1.2, 1.2)
                plt.bar(np.arange(3), [a[k, i, j]/max_a, mu[k, i, j]/max_mu, log_sigma[k, i, j]/max_log_sigma], color=coef_colors)
                ax.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
                
        fig.tight_layout(pad=0.0)
        plt.subplots_adjust(bottom=0.0, left=0.0, right=1.0, top=1.0)

        if t == 1:
            plt.show()
        else:
            plt.savefig("plot_coefs_{}.png".format(k))
            plt.close()
    
    if t > 1:
        os.system("ffmpeg -r 5 -i plot_coefs_%d.png plot_coefs.mp4")
```

```python
alpha = 20
phis = torch.ones(1, 1)
stride = 1

viz_phis(phis, alpha)

a = torch.tensor([[0], [0], [0.5], [0], [1], [0.5], [0.5], [0], [0]])
mu = torch.zeros_like(a)
sigma = torch.ones_like(a)

log_sigma = torch.log(sigma)
plot_coefs(a, mu, log_sigma)

Y = get_Y(alpha, stride, phis,
      a, mu, sigma,
      plot=True)
```

```python
## 1d_warm_start_each_temporal_smott

import tqdm

a_t = []
mu_t = []
log_sigma_t = []

alpha = 4
phis = torch.tensor([[0.33, 0.67], [0.67, 0.33]])
colors = ['b', 'r']
stride = 1

a = torch.zeros(4, 2, requires_grad=True)
mu = torch.zeros(4, 2, requires_grad=True)
log_sigma = torch.zeros(4, 2, requires_grad=True)

for t in tqdm.tqdm(range(17)):
    target = torch.zeros(33)
    for j in range(4, 8+t):
        target[j] = 1/(4+t)*(j-4)
    for j in range(8+t, 28):
        target[j] = 1 - 1/(20-t)*(j-(8+t))
    plt.plot(target, color="green")
    plt.show()

    optimizer = torch.optim.Adam([a, mu, log_sigma], lr=0.1)

    for itr in range(20):
        optimizer.zero_grad()

        sigma = torch.exp(log_sigma)
        Y = get_Y(alpha, stride, phis,
          a, mu, sigma,
          plot=False)
        if t == 0:
            nll = (target - Y).abs().mean() \
                + 0.1*(a.abs().mean() + mu.abs().mean() + log_sigma.abs().mean())
        else:
            nll = (target - Y).abs().mean() \
                + 0.1*(a.abs().mean() + mu.abs().mean() + log_sigma.abs().mean() \
                      + (a-a_t[-1]).abs().mean() + (mu-mu_t[-1]).abs().mean() + (log_sigma-log_sigma_t[-1]).abs().mean())
        nll.backward()

        optimizer.step()

    Y = get_Y(alpha, stride, phis,
      a, mu, sigma,
      plot=True, colors=colors)

    a_t.append(a.detach().clone())
    mu_t.append(mu.detach().clone())
    log_sigma_t.append(log_sigma.detach().clone())
        
plot_coefs(torch.stack(a_t), torch.stack(mu_t), torch.stack(log_sigma_t))
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

def gauss2d(mu_x, mu_y, sigma_x, sigma_y, rot):
    Lambda = torch.eye(2)
    Lambda[0, 0] *= sigma_x**2
    Lambda[1, 1] *= sigma_y**2

    U = torch.ones(2, 2)
    U[0, 0] *= torch.cos(rot)
    U[0, 1] *= -torch.sin(rot)
    U[1, 0] *= torch.sin(rot)
    U[1, 1] *= torch.cos(rot)

    Sigma = torch.matmul(U, torch.matmul(Lambda, U.T))

    mu = torch.ones(2)
    mu[0] *= mu_x
    mu[1] *= mu_y
    
    distr = torch.distributions.MultivariateNormal(mu, Sigma)

    k = torch.exp(distr.log_prob(mu))

    def f(x, y):
        return torch.exp(distr.log_prob(torch.tensor([x, y]).float())) / k

    return f

def get_Z(
    alpha_x, alpha_y, stride_x, stride_y, phis,
    a, mu_x, mu_y, sigma_x, sigma_y, rot,
    plot=False, c=None,
):
    n_y = len(a)
    n_x = len(a[0])
    philter_len_y = len(phis[0][0])
    philter_len_x = len(phis[0][0][0])
    N_y = int(np.ceil(1 + alpha_y * philter_len_y * (1 + (n_y-1)*stride_y)))
    N_x = int(np.ceil(1 + alpha_x * philter_len_x * (1 + (n_x-1)*stride_x)))

    if plot:
        fig = plt.figure(figsize=(4, 8))
        ax = fig.add_subplot(2, 1, 1, projection='3d')
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.set_zlabel("z")
        
        Y, X = torch.meshgrid(torch.arange(N_y), torch.arange(N_x))

    Z = torch.zeros(N_y, N_x)
    for i_y in range(n_y):
        for i_x in range(n_x):
            for k_y in range(philter_len_y):
                for k_x in range(philter_len_x):
                    this_rot = rot[i_y, i_x]
                    y_offset = -sigma_y[i_y, i_x]*alpha_y*(philter_len_y/2 - (k_y+0.5))
                    x_offset = -sigma_x[i_y, i_x]*alpha_x*(philter_len_x/2 - (k_x+0.5))
                    this_mu_y = alpha_y*philter_len_y*stride_y*(i_y + mu_y[i_y, i_x]) \
                        + alpha_y*philter_len_y/2 \
                        + y_offset*torch.cos(this_rot) + x_offset*torch.sin(this_rot)
                    this_mu_x = alpha_x*philter_len_x*stride_x*(i_x + mu_x[i_y, i_x]) \
                        + alpha_x*philter_len_x/2 \
                        + x_offset*torch.cos(this_rot) - y_offset*torch.sin(this_rot)
                    this_sigma_y = sigma_y[i_y, i_x]*alpha_y/2
                    this_sigma_x = sigma_x[i_y, i_x]*alpha_x/2
                    this_gauss2d = gauss2d(this_mu_x, this_mu_y, this_sigma_x, this_sigma_y, this_rot)
                    
                    min_y = int(max(0, this_mu_y.detach() \
                                    -torch.cos(this_rot.detach())*(3*this_sigma_y.detach()) \
                                    -torch.abs(torch.sin(this_rot.detach()))*(3*this_sigma_x.detach())))
                    max_y = int(min(N_y, this_mu_y.detach() \
                                    +torch.cos(this_rot.detach())*(3*this_sigma_y.detach()) \
                                    +torch.abs(torch.sin(this_rot.detach()))*(3*this_sigma_x.detach())))
                    min_x = int(max(0, this_mu_x.detach() \
                                    -torch.cos(this_rot.detach())*(3*this_sigma_x.detach()) \
                                    -torch.abs(torch.sin(this_rot.detach()))*(3*this_sigma_y.detach())))
                    max_x = int(min(N_x, this_mu_x.detach() \
                                    +torch.cos(this_rot.detach())*(3*this_sigma_x.detach()) \
                                    +torch.abs(torch.sin(this_rot.detach()))*(3*this_sigma_y.detach())))
                    
                    this_Z = torch.zeros_like(Z)
                    for j_y in range(min_y, max_y):
                        for j_x in range(min_x, max_x):
                            this_Z[j_y, j_x] = a[i_y, i_x] * phis[i_y][i_x][k_y][k_x] * this_gauss2d(j_x, j_y)
                    Z += this_Z

                    if plot and c is not None:
                        ax.plot_surface(X, Y, this_Z.detach(), color=c[i_y][i_x], alpha=0.5)

    if plot:
        ax.plot_surface(X, Y, Z.detach(), color="black", alpha=0.2)

        ax = fig.add_subplot(2, 1, 2)
        ax.imshow(Z.detach().numpy()[::-1], cmap="gray", vmin=0, vmax=1)

        plt.show()

    return Z

def viz_phis2d(phis, alpha_x, alpha_y, c=None):
    N_y = len(phis)
    N_x = len(phis[0])
    philter_len_y = len(phis[0][0])
    philter_len_x = len(phis[0][0][0])
    
    Y, X = torch.meshgrid(torch.arange(1 + 2*alpha_y), torch.arange(1 + 2*alpha_x))
    
    if c is None:
        c = [["black" for _ in range(N_x)] for _ in range(N_y)]

    fig = plt.figure(figsize=(2*N_x, 2*N_y*3))
    
    for i_y in range(N_y):
        for i_x in range(N_x):
            ax = fig.add_subplot(3*N_y, N_x, N_x*(3*i_y) + i_x + 1)
            ax.imshow(np.array(phis[i_y][i_x].detach().numpy()[::-1]), cmap="gray", vmin=0, vmax=1)

            ax = fig.add_subplot(3*N_y, N_x, N_x*(3*i_y+1) + i_x + 1, projection="3d")
            Z = get_Z(alpha_x, alpha_y, stride_x=1, stride_y=1, phis=[[phis[i_y][i_x]]],
                a=torch.ones(1, 1), mu_x=torch.zeros(1, 1), mu_y=torch.zeros(1, 1),
                    sigma_x=torch.ones(1, 1), sigma_y=torch.ones(1, 1), rot=torch.zeros(1, 1))
            ax.plot_surface(X, Y, Z.detach(), color=c[i_y][i_x], alpha=0.2)

            ax = fig.add_subplot(3*N_y, N_x, N_x*(3*i_y+2) + i_x + 1)
            ax.imshow(Z.numpy()[::-1], cmap="gray", vmin=0, vmax=1)
            
    plt.show()

def plot_coefs2d(a, mu_x, mu_y, log_sigma_x, log_sigma_y, rot, c):
    coef_names = ["a", "mu_x", "mu_y", "log_sigma_x", "log_sigma_y", "rot"]
    N_y = len(a)
    N_x = len(a[0])

    fig = plt.figure(figsize=(2, 12))
    
    for (j, x) in enumerate([a, mu_x, mu_y, log_sigma_x, log_sigma_y, rot]):
        x = x.detach().numpy()
        ax = fig.add_subplot(6, 1, j+1, projection="3d")
        for i_y in range(N_y):
            for i_x in range(N_x):
                ax.plot([i_x-0.25, i_x+0.25], [i_y-0.25, i_y+0.25], [0, 0], c=c[i_y][i_x])
                ax.plot([i_x+0.25, i_x-0.25], [i_y-0.25, i_y+0.25], [0, 0], c=c[i_y][i_x])
                ax.plot([i_x-0.25, i_x+0.25], [i_y-0.25, i_y+0.25], [x[i_y, i_x], x[i_y, i_x]], c=c[i_y][i_x])
                ax.plot([i_x+0.25, i_x-0.25], [i_y-0.25, i_y+0.25], [x[i_y, i_x], x[i_y, i_x]], c=c[i_y][i_x])
                ax.plot([i_x, i_x], [i_y, i_y], [0, x[i_y, i_x]], c=c[i_y][i_x])
        ax.set_title(coef_names[j])

    plt.show()
```
